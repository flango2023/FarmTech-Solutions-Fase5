## 1. Dashboard de KPIs - Sistema CultureX

```python
# KPIs do sistema
kpis = {
    'Integração do Sistema': {'valor': 96.7, 'meta': 60, 'unidade': '%'},
    'Precisão ML': {'valor': 94.7, 'meta': 90, 'unidade': '%'},
    'Latência': {'valor': 87, 'meta': 100, 'unidade': 'ms'},
    'Disponibilidade': {'valor': 99.7, 'meta': 99, 'unidade': '%'},
    'Leituras/Hora': {'valor': 150, 'meta': 100, 'unidade': 'readings'},
    'Eventos Processados': {'valor': 45, 'meta': 30, 'unidade': 'events/hour'}
}

# Dashboard de KPIs
fig = make_subplots(
    rows=2, cols=3,
    specs=[[{"type": "indicator"}, {"type": "indicator"}, {"type": "indicator"}],
           [{"type": "indicator"}, {"type": "indicator"}, {"type": "indicator"}]],
    subplot_titles=list(kpis.keys())
)

positions = [(1,1), (1,2), (1,3), (2,1), (2,2), (2,3)]

for i, (kpi_name, kpi_data) in enumerate(kpis.items()):
    row, col = positions[i]
    
    color = "green" if kpi_data['valor'] >= kpi_data['meta'] else "red"
    if kpi_name == 'Latência':
        color = "green" if kpi_data['valor'] <= kpi_data['meta'] else "red"
    
    fig.add_trace(
        go.Indicator(
            mode="number+delta",
            value=kpi_data['valor'],
            delta={'reference': kpi_data['meta']},
            title={'text': f"{kpi_name}<br><span style='font-size:0.8em;color:gray'>{kpi_data['unidade']}</span>"},
            number={'font': {'color': color}}
        ),
        row=row, col=col
    )

fig.update_layout(
    title="KPIs do Sistema CultureX - Sprint 3",
    height=600,
    width=1000
)

fig.show()
```

## 2. Gauge de Integração do Sistema

```python
# Gauge de Integração
fig = go.Figure(go.Indicator(
    mode="gauge+number+delta",
    value=96.7,
    domain={'x': [0, 1], 'y': [0, 1]},
    title={'text': "Nível de Integração do Sistema (%)"},
    delta={'reference': 60},
    gauge={
        'axis': {'range': [None, 100]},
        'bar': {'color': "darkblue"},
        'steps': [
            {'range': [0, 60], 'color': "lightgray"},
            {'range': [60, 90], 'color': "yellow"},
            {'range': [90, 100], 'color': "green"}
        ],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': 90
        }
    }
))

fig.update_layout(
    title="Integração do Sistema CultureX - Meta: 60% | Alcançado: 96.7%",
    height=500,
    width=600
)

fig.show()
```

## 3. Performance dos Modelos ML

```python
# Gráfico radar para métricas de classificação
categories = ['Precisão', 'Recall', 'F1-Score', 'Acurácia']
values = [94.7, 95.6, 93.9, 94.7]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=values,
    theta=categories,
    fill='toself',
    name='Classificador de Interações',
    line_color='#FF6B6B'
))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[80, 100]
        )),
    showlegend=True,
    title="Performance do Modelo de Classificação ML"
)

fig.show()
```

## 4. Timeline de Atividade dos Sensores

```python
# Simular dados de performance
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=30, freq='D')

performance_data = pd.DataFrame({
    'Data': dates,
    'Leituras_Sensores': np.random.poisson(150, 30),
    'Eventos_Processados': np.random.poisson(45, 30),
    'Precisao_ML': np.random.normal(94.7, 1.2, 30),
    'Latencia_ms': np.random.normal(87, 8, 30)
})

# Timeline de leituras
fig = px.line(
    performance_data,
    x='Data',
    y='Leituras_Sensores',
    title='Timeline de Atividade dos Sensores IoT',
    labels={'Leituras_Sensores': 'Número de Leituras'}
)

fig.add_hline(
    y=performance_data['Leituras_Sensores'].mean(),
    line_dash="dash",
    annotation_text=f"Média: {performance_data['Leituras_Sensores'].mean():.1f}"
)

fig.show()
```

## 5. Mapa de Calor - Atividade por Sensor

```python
# Simular dados de sensores
sensor_types = ['Temperatura', 'Umidade', 'Luz', 'Toque', 'Presença', 'Proximidade']
locations = ['Hall Principal', 'Painel Interativo', 'Área Exposição', 'Ambiente']

# Gerar dados aleatórios para o heatmap
np.random.seed(42)
heatmap_data = np.random.randint(20, 100, (len(locations), len(sensor_types)))

fig = px.imshow(
    heatmap_data,
    x=sensor_types,
    y=locations,
    color_continuous_scale='Viridis',
    title='Mapa de Calor - Atividade dos Sensores por Localização',
    aspect='auto'
)

fig.update_layout(
    xaxis_title='Tipo de Sensor',
    yaxis_title='Localização',
    width=800,
    height=500
)

fig.show()
```

## 6. Dashboard Completo de Performance

```python
# Dashboard com múltiplos gráficos
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Leituras de Sensores', 'Eventos Processados', 
                   'Precisão ML (%)', 'Latência (ms)'),
    specs=[[{"secondary_y": False}, {"secondary_y": False}],
           [{"secondary_y": False}, {"secondary_y": False}]]
)

# Gráfico 1: Leituras de Sensores
fig.add_trace(
    go.Scatter(x=performance_data['Data'], y=performance_data['Leituras_Sensores'],
               mode='lines+markers', name='Leituras', line=dict(color='#FF6B6B')),
    row=1, col=1
)

# Gráfico 2: Eventos Processados
fig.add_trace(
    go.Bar(x=performance_data['Data'], y=performance_data['Eventos_Processados'],
           name='Eventos', marker_color='#4ECDC4'),
    row=1, col=2
)

# Gráfico 3: Precisão ML
fig.add_trace(
    go.Scatter(x=performance_data['Data'], y=performance_data['Precisao_ML'],
               mode='lines', name='Precisão ML', line=dict(color='#96CEB4')),
    row=2, col=1
)

# Gráfico 4: Latência
fig.add_trace(
    go.Scatter(x=performance_data['Data'], y=performance_data['Latencia_ms'],
               mode='markers', name='Latência', marker=dict(color='#FFEAA7')),
    row=2, col=2
)

fig.update_layout(
    title="Dashboard de Performance - Sistema CultureX",
    showlegend=False,
    height=800,
    width=1200
)

fig.show()
```
